# Part1

```python
features = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
targets = df['target']
```

```python
train_features = features.iloc[:149].copy()
ptest = features.iloc[149].copy()
true_class = targets.iloc[149]
train_targets = targets.iloc[:149]
ptest
```

```python
distances_no_norm = np.linalg.norm(train_features.values - ptest.values, ord=2, axis=1)
# distances = np.sqrt(np.sum((train_features - ptest)**2, axis=1))
distances_no_norm
```

```python
nearest_indices_no_norm = np.argsort(distances_no_norm)[:3]
print("nearest_indices_no_norm:", nearest_indices_no_norm)
neighbors_classes_no_norm = train_targets.iloc[nearest_indices_no_norm].values
print("neighbors_classes_no_norm :", neighbors_classes_no_norm)
counts_no_norm = np.bincount(neighbors_classes_no_norm.astype(int))
print("counts_no_norm :" , counts_no_norm)
pred_class_no_norm = np.argmax(counts_no_norm)
print("pred_class_no_norm :" ,pred_class_no_norm)
correct_no_norm = pred_class_no_norm == true_class
```

```python
for idx in nearest_indices_no_norm:
    features_row = train_features.iloc[idx]
    class_row = train_targets.iloc[idx]
    dist_val = distances_no_norm[idx]
    print(f"Index {idx}: SL={features_row['sepal length (cm)']:.2f}, SW={features_row['sepal width (cm)']:.2f}, PL={features_row['petal length (cm)']:.2f}, PW={features_row['petal width (cm)']:.2f}, Class={class_row}, Distance={dist_val:.4f}")
```

```python
print(f"پیش‌بینی k=3 (بدون نرمال): {pred_class_no_norm} (درست: {correct_no_norm})")
```

# Normazied data:

```python
mu = train_features.mean()
sigma = train_features.std()
train_features_z = (train_features - mu) / sigma
ptest_z = (ptest - mu) / sigma
```

```python
#distances_z = np.linalg.norm(train_features_z.values - ptest_z.values, ord=2, axis=1)
distances = np.sqrt(np.sum((train_features - ptest_z)**2, axis=1))
distances_z
```

```python
nearest_indices_z = np.argsort(distances_z)[:3]
print("nearest_indices_z :" ,nearest_indices_z)
neighbors_classes_z = train_targets.iloc[nearest_indices_z].values
print("neighbors_classes_z :" , neighbors_classes_z)
counts_z = np.bincount(neighbors_classes_z.astype(int))
print("counts_z :" , counts_z)
pred_class_z = np.argmax(counts_z)
print("pred_class_z :" , pred_class_z)
correct_z = pred_class_z == true_class
```

```python
for idx in nearest_indices_z:
    features_row = train_features.iloc[idx]
    class_row = train_targets.iloc[idx]
    dist_val = distances_z[idx]
    print(f"Index {idx}: SL={features_row['sepal length (cm)']:.2f}, SW={features_row['sepal width (cm)']:.2f}, PL={features_row['petal length (cm)']:.2f}, PW={features_row['petal width (cm)']:.2f}, Class={class_row}, Z-Distance={dist_val:.4f}")
```

```python
print(f"پیش‌بینی k=3 (Z-score دستی): {pred_class_z} (درست: {correct_z})")
```
